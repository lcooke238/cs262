```python
import os

paths = ["../logs/experiment_1/log_0.txt", "../logs/experiment_1/log_1.txt", "../logs/experiment_1/log_2.txt"]

#logical clocks
def logical_clock_analysis(logfilepath):
    #get logical clock numbers from each log file
    log_clock_nums = []
    with open(logfilepath, "r") as file:
        data = file.readlines()
        for line in data:
            a = line.split(" ")
            if a[0] != "LOG":
                log_clock_nums.append(int(a[0]))
            else:
                print(line)
    # print(log_clock_nums)

    #get the difference between subsequent clock vals
    difs = []
    for i in range(len(log_clock_nums)-1):
        difs.append(log_clock_nums[i+1] - log_clock_nums[i])
    # print(difs)

    #get the average, max, and min of the differences
    print(f"average: {sum(difs)/len(difs)}")
    print(f"min: {min(difs)}")
    print(f"max: {max(difs)}")


#value drift
def value_drift_analysis(logfilepath):
    # get logical time diffs
    log_clock_nums = []
    sys_clock_nums = []
    clock_difs = []
    with open(logfilepath, "r") as file:
        data = file.readlines()
        for line in data:
            a = line.split(" ")
            if a[0] != "LOG":
                log_clock_nums.append(int(a[0]))
                sys_clock_nums.append(float(a[2][:len(a[2])-1]))
                clock_difs.append(float(a[2][:len(a[2])-1])-int(a[0]))
            else:
                print(line)

    #get the difference between subsequent clock vals
    log_difs = []
    sys_difs = []
    difs = []
    for i in range(len(log_clock_nums)-1):
        log_difs.append(log_clock_nums[i+1] - log_clock_nums[i])
        sys_difs.append(sys_clock_nums[i+1] - sys_clock_nums[i])
        difs.append((clock_difs[i+1] - clock_difs[i]))#/(log_clock_nums[i+1] - log_clock_nums[i]))
    
    # compare the diffs for drift
    # print(log_difs)
    # print(sys_difs)
    # print(difs)
    print(f"max: {max(difs)}")
    print(f"min: {min(difs)}")
    print(f"avg: {sum(difs)/len(difs)}")
```

```python
#run logical clock analysis
for path in paths:
    logical_clock_analysis(path)
    print()
    print()

#run value drift analysis
for path in paths:
    value_drift_analysis(path)
```
