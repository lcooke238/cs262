## Numerical Analysis

```python
import os

#logical clocks
def logical_clock_analysis(logfilepath):
    #get logical clock numbers from each log file
    log_clock_nums = []
    with open(logfilepath, "r") as file:
        data = file.readlines()
        for line in data:
            a = line.split(" ")
            if a[0] != "LOG":
                log_clock_nums.append(int(a[0]))
            else:
                print(line)
    # print(log_clock_nums)

    #get the difference between subsequent clock vals
    difs = []
    for i in range(len(log_clock_nums)-1):
        difs.append(log_clock_nums[i+1] - log_clock_nums[i])
    # print(difs)

    #get the average, max, and min of the differences
    print(f"average: {sum(difs)/len(difs)}")
    print(f"min: {min(difs)}")
    print(f"max: {max(difs)}")


#value drift
def value_drift_analysis(logfilepath):
    # get logical time diffs
    log_clock_nums = []
    sys_clock_nums = []
    clock_difs = []
    with open(logfilepath, "r") as file:
        data = file.readlines()
        for line in data:
            a = line.split(" ")
            if a[0] != "LOG":
                log_clock_nums.append(int(a[0]))
                sys_clock_nums.append(float(a[2][:len(a[2])-1]))
                clock_difs.append(float(a[2][:len(a[2])-1])-int(a[0]))
            else:
                print(line)

    #get the difference between subsequent clock vals
    log_difs = []
    sys_difs = []
    difs = []
    for i in range(len(log_clock_nums)-1):
        log_difs.append(log_clock_nums[i+1] - log_clock_nums[i])
        sys_difs.append(sys_clock_nums[i+1] - sys_clock_nums[i])
        difs.append((clock_difs[i+1] - clock_difs[i]))#/(log_clock_nums[i+1] - log_clock_nums[i]))
    
    # compare the diffs for drift
    # print(log_difs)
    # print(sys_difs)
    # print(difs)
    print(f"max: {max(difs)}")
    print(f"min: {min(difs)}")
    print(f"avg: {sum(difs)/len(difs)}")
```

```python
paths = ["../logs/experiment_3/log_0.txt", "../logs/experiment_3/log_1.txt", "../logs/experiment_3/log_2.txt"]

#run logical clock analysis
for path in paths:
    logical_clock_analysis(path)
    print()
    print()

#run value drift analysis
for path in paths:
    value_drift_analysis(path)
```

## Visual Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

def logical_clock_pull(logfilepath):
    #gets just logical clock numbers from each log file
    log_clock_nums = []
    with open(logfilepath, "r") as file:
        data = file.readlines()
        for line in data:
            a = line.split(" ")
            if a[0] != "LOG":
                log_clock_nums.append(int(a[0]))
    return log_clock_nums

def clock_rate_pull(logfilepath):
    #gets just clock rates from each log file
    with open(logfilepath, "r") as file:
        line = file.readline()
        return int(line.split(" ")[-1:][0]) 
```

```python
for i in range(1, 6):
    paths = [f"../logs/experiment_{i}/log_0.txt", f"../logs/experiment_{i}/log_1.txt", f"../logs/experiment_{i}/log_2.txt"]
    
    clock_rates = []
    for path in paths:
        clock_rates.append(clock_rate_pull(path))

    # num of seconds elapsed
    times = [[elt / clock_rate for elt in range(clock_rate * 60)] for clock_rate in clock_rates]
    clock_times = []
    for j in range(3):
        clock_times.append(logical_clock_pull(paths[j])[:(clock_rates[j] * 60)])

    fig, axis = plt.subplots(1, 1, figsize=(8, 8))

    for j in range(3):
        sns.lineplot(x=times[j], y=clock_times[j], ax=axis, label=f'machine {j} with clock rate {clock_rates[j]}')

    axis.set_xlabel('seconds elapsed',fontsize=12)
    axis.set_ylabel('local logical clock time',fontsize=12)
    axis.set_title(f'synchronicity of three machines for experiment {i}',fontsize=15)
    axis.legend()

    plt.show()
```
